```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time
from config import usuario, senha

# Define as opções do navegador
options = webdriver.FirefoxOptions()

# options.headless = True  # Define como True para rodar o Firefox sem abrir a janela do navegador

# Inicializa o navegador
browser = webdriver.Firefox(options=options)

# Faz login no site
url_login = 'https://intranet.policiamilitar.mg.gov.br'
browser.get(url_login)

time.sleep(2)  # Espera um tempo para que a página carregue completamente

# Preenche os campos de login e senha
input_usuario = browser.find_element('id', 'textLogin')
input_usuario.send_keys(usuario)

input_senha = browser.find_element('id', 'senha')
input_senha.send_keys(senha)

# Submete o formulário de login
input_senha.send_keys(Keys.ENTER)

time.sleep(2)  # Espera um tempo para que a página carregue completamente

df = pd.DataFrame(columns=['Nº POL', 'NOME', 'LOTAÇÃO'])
dados = []
for p in range(1, 15):
    url_dados = f'https://intranet.policiamilitar.mg.gov.br/lite/recursos-humanos/web/?p={p}&nome=&nr_pol=&postograd=&genero=T&ativo=A&unidade=2560&municipio=&bairro=&endereco='

    # Acessa a página de dados
    browser.get(url_dados)
    time.sleep(2)  # Espera um tempo para que a página carregue completamente
    
#     Faz o parsing do HTML do site com o Beautiful Soup
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Encontra todas as divs com a class "item busca-pessoal"
    containers = soup.find_all(class_='item busca-pessoal')
#     if len(containers) == 0:
#         containers = soup.find_all('a', {'class': 'item busca-pessoal'})
    
#     divs2 = soup.find_all('div')

    # Extrai os dados de cada registro e adiciona a um dataframe do Pandas
    
    for container in containers:
        dt = container.find('dt').text.strip()
        nome = container.find_all('dd')[0].text.strip()
        lotacao = container.find_all('dd')[1].text.strip()
        dados.append({'Nº POL': dt, 'NOME': nome, 'LOTAÇÃO': lotacao})
    time.sleep(2)

for dicionario in dados:
    df = df.append(dicionario, ignore_index=True)
df.to_excel('dados2.xlsx', index=False)
        
# Fecha o navegador    
browser.quit()
```
